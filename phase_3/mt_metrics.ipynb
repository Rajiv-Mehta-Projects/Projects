```python
import json
import matplotlib.pyplot as plt
```

```python
with open('results/100_dual.json', 'r') as f:
    loaded = json.load(f)
dual = loaded['accuracy']

with open('results/100_all_in_one.json', 'r') as f:
    loaded = json.load(f)
single = loaded['accuracy']
```

```python
plt.plot(dual[0], label='client 1')
plt.plot(dual[1], label='client 2')
plt.title("Dual input")
plt.xlabel("Iteration")
plt.ylabel("Accuracy")
plt.legend()
plt.show()
```

```python
plt.plot(single[0], label='client 1')
plt.plot(single[1], label='client 2')
plt.title("SImple input (One for all)")
plt.xlabel("Iteration")
plt.ylabel("Accuracy")
plt.ylim(top=1.0, bottom=0.47)
plt.legend()
plt.show()
```
